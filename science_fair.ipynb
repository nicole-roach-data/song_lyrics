```python
# Initialize spacy 'en' model, keeping only tagger component (for efficiency)

nlp = spacy.load('en', disable=['parser', 'ner'])
#named entity recognition was used in the model for better results but takes a long time to run
```

```python
#Tokenize & Clean
#Step 1
def sent_to_words(sentences):
    yield(gensim.utils.simple_preprocess(str(sentences), deacc=True))  # deacc=True removes punctuations
```

```python
#Step 2.5
#Custom Stop Words
stop = list(text.ENGLISH_STOP_WORDS)
new_words = ["\'ll", 'im', 've', 'll', 'ah', 'na', 'hey', 'oh', 'em', 'hoo', 'ay', 'eh', 'ho', 'huh', 'uh', 'ooh', 'dum', 'woo', 'eh', 'woah', 'la', 'people', 'yeahhahahhhahhhhhahhh', 'oohhh', 'yaaaaa', 'don', '12', '10']
stop_words = text.ENGLISH_STOP_WORDS.union(new_words)
```

```python
#Step 2
#Lemmatize
def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append(" ".join([token.lemma_ if token.lemma_ not in ['-PRON-'] else '' for token in doc if token.pos_ in allowed_postags]))
    return texts_out
```

```python
#Step 3
#Create Doc Word Matrix
vectorizer = CountVectorizer(analyzer='word',       
                             min_df=1,
                             max_df = 1,                                # minimum reqd occurences of a word 
                             stop_words = stop_words,             # remove stop words
                             lowercase=True,                   # convert all words to lowercase
                             token_pattern='[a-zA-Z0-9]{3,}',  # num chars > 3
                            )
```
